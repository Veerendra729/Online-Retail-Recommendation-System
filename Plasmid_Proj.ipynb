```python
import pandas as pd

# Load the uploaded Excel file
df = pd.read_excel("OnlineRetail (1).xlsx")

# Preview first 5 rows
df.head()
```

```python
# Remove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Keep only valid transactions
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Add total price column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
import matplotlib.pyplot as plt

# Group by product and get top 10
top_products = (
    df.groupby('Description')['Quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Show result
print("Top 10 Recommended Products:\n")
print(top_products)

# Plot
top_products.plot(kind='barh', figsize=(10,6), title='Top 10 Most Purchased Products')
plt.xlabel("Quantity Sold")
plt.tight_layout()
plt.show()


```

```python
# Convert InvoiceDate to datetime (if not already done)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Set InvoiceDate as index
df.set_index('InvoiceDate', inplace=True)

# Group by month and sum total price
monthly_sales = df['TotalPrice'].resample('M').sum()

# Plot monthly trend
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
monthly_sales.plot(title='Figure 4: Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales (£)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Reset index if you ran the previous cell
df.reset_index(inplace=True)

# Group by Country and count unique invoices
country_sales = df.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False)

# Optionally exclude United Kingdom (most dominant)
country_sales = country_sales[country_sales.index != 'United Kingdom'].head(10)

# Plot bar chart
country_sales.plot(kind='barh', figsize=(10,6), color='skyblue', title='Figure 5: Top 10 Countries by Transaction Volume')
plt.xlabel('Number of Transactions')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
